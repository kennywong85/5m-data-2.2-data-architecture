### Question 1

Question: How would you create a new hash in Redis to store information about a person, with `john_doe` as the key and include the name (`John Doe`), age (`35`), and email address (`john@email.com`)?

Answer:

```python
# Redis creates a little profile card called john_doe. Inside that card, you store name, age, and email. 
# Then hgetall() opens the card and shows you what is inside.

# r.hset(...) = creates/updates a Redis hash.
# "john_doe" = the Redis key, like the name of the mini-profile.
# mapping={...} = the fields stored inside that hash.
r.hset(
    "john_doe",
    mapping={
        "name": "John Doe",
        "age": 35,
        "email": "john@email.com",
    }
)

# r.hgetall("john_doe") = reads back the whole hash so you can check it worked
r.hgetall("john_doe")
```

Answer:

```python
# You are opening the public Landsat cloud cupboard, looking at the files inside, 
# and printing only the first 10 file labels with their sizes. 
# The break stops your notebook from vomiting the entire satellite archive into your face

from google.cloud import storage

client = storage.Client.create_anonymous_client()
bucket = client.get_bucket("gcp-public-data-landsat")

# gets the objects/files inside the bucket.
blobs = bucket.list_blobs()

# loops through the blobs and gives each one a running number starting from 1.
for i, blob in enumerate(blobs, start=1):
    # blob.name = the object/blob filename.
    # blob.size = the object size in bytes.
    print(f"{i}. {blob.name} - {blob.size} bytes")

    # if i == 10: break = stop after printing 10 objects.
    if i == 10:
        break
```
